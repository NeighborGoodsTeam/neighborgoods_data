```python
import requests
import pandas as pd
import functions
import json

from bs4 import BeautifulSoup
from googlemaps import Client
```

```python
res_list = [requests.get('https://decemberthieves.com/collections/shop-all?page=' + str(i)) for i in range(1, 15)]
```

```python
soup_list = [BeautifulSoup(res.content) for res in res_list]
```

```python
item_lists = [[h2.find(name='a').text for h2 in soup.find_all(name='h2', attrs={'class' : 'ProductItem__Title Heading'})] for soup in soup_list]
```

```python
item_lists
```

```python
catalog = []

for item_list in item_lists:
    for item in item_list:
        catalog.append(item)
```

```python
with open('../google_key.txt', 'r') as f:
    google_key = f.read()

google_api = Client(google_key)
```

```python
place_search = google_api.places(
    query='december',
    location=(42.35773674139874, -71.07001746015413),
    type='establishment'
)
```

```python
place_result = place_search['results'][0]
```

```python
place_entry = functions.entry_from_google(place_result)
```

```python
place_entry['products'] = catalog
```
