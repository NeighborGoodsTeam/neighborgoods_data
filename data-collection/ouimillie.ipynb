```python
res_apparel = [requests.get('https://ouimillie.com/collections/apparel?page=' + str(i)) for i in range(1, 18)]
```

```python
soup_apparel = [BeautifulSoup(res.content) for res in res_apparel]
```

```python
apparel_lists = [[p.text for p in soup.find_all(name='p', attrs={'class' : 'grid-link__title'})] for soup in soup_apparel]
```

```python
full_apparel = []

for apparel_list in apparel_lists:
    for item in apparel_list:
        full_apparel.append(item)
```
