```python
import requests
import pandas as pd
import functions
import json

from bs4 import BeautifulSoup
from googlemaps import Client
```

```python
res = requests.get('https://loyalsupplyco.com/collections/all-items')
```

```python
soup = BeautifulSoup(res.content)
```

```python
loyal_h3_list = soup.find_all('h3', {'class' : 'product-list-item-title'})
```

```python
first = loyal_h3_list[0]
```

```python
first.text
```

```python
loyal_items = [h.text for h in loyal_h3_list]
```

```python
loyal_items
```

```python
with open('google_key.txt', 'r') as f:
    google_key = f.read()

google_api = Client(google_key)
```

```python
test_search = google_api.places(
    query='loyal',
    location=(42.38160748236796, -71.08542406272488),
    type='establishment'
)
```

```python
loyal_supply = test_search['results'][0]
```

```python
loyal_entry = functions.entry_from_google(loyal_supply)
```

```python
loyal_entry['products'] = loyal_items
```

```python
loyal_entry['email'] = 'info@loyalsupplyco.com'
loyal_entry['url'] = 'https://loyalsupplyco.com/'
loyal_entry['description'] = 'This hip, curated shop offers unique gifts & home goods, from artisanal wool blankets to beard soap.'
loyal_entry['hours'] = {
    'Monday' : 'Closed',
    'Tuesday' : 'Closed',
    'Wednesday' : 'Closed',
    'Thursday' : 'Closed',
    'Friday' : 'Closed',
    'Saturday' : (10, 18),
    'Sunday' : (12, 17)
}
```
