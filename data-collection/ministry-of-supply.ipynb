```python
import requests
import pandas as pd
import functions
import json

from bs4 import BeautifulSoup
from googlemaps import Client
```

```python
res_men = requests.get('https://www.ministryofsupply.com/men/shop-all')
```

```python
soup_men = BeautifulSoup(res_men.content)
```

```python
items_men = [span.text for span in soup_men.find_all(name='span', attrs={'class' : 'CardProduct__collection t-h4'})]
```

```python
res_women = requests.get('https://www.ministryofsupply.com/women/shop-all')
```

```python
soup_women = BeautifulSoup(res_women.content)
```

```python
items_women = [span.text for span in soup_women.find_all(name='span', attrs={'class' : 'CardProduct__collection t-h4'})]
```

```python
res_mask = requests.get('https://www.ministryofsupply.com/all/mask-program')
```

```python
soup_mask = BeautifulSoup(res_mask.content)
```

```python
items_mask = [span.text for span in soup_mask.find_all(name='span', attrs={'class' : 'CardProduct__collection t-h4'})]
```

```python
item_lists = [items_men, items_women, items_mask]
```

```python
catalog = []

for item_list in item_lists:
    for item in item_list:
        catalog.append(item)
```

```python
with open('./google_key.txt', 'r') as f:
    google_key = f.read()

google_api = Client(google_key)
```

```python
place_search = google_api.places(
    query='ministry',
    location=(42.34957529104881, -71.08383778182603),
    type='establishment'
)
```

```python
place_result = place_search['results'][0]
```

```python
place_entry = functions.entry_from_google(place_result)
```

```python
place_entry['products'] = catalog
```
